# Split data into train test split

dividing data into predictors and target
Target is what we want to predict given the predictors

```python
target = data['Strength']
data_columns = data.columns
predictors = data[data_columns[data_columns!='Strength']]
```

```python
predictors.head()
```

```python
target.head()
```

# Normalizing the Data

subtracting mean and dividing by the standard deviation

```python
norm_data = (predictors - predictors.mean())/ predictors.std()
```

```python
#Doing EDA
#Checking the data to understand it better
norm_data.tail()
```

```python
norm_data.describe() # seeing the statistics of the data
```

```python
norm_data.shape
```

```python
num_col = norm_data.shape[1] #saving the number of predictors to num_col
```

# Importing Libraries necessary for model building

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# Building the neural network!

Building the function for regression

```python
my_mse = []

for i in range(50):
    x_train,x_test,y_train,y_test = train_test_split(norm_data,target,test_size=0.30,random_state = 0)
    model = Sequential()
    #model layers
    model.add(Dense(10,activation= 'relu',input_shape = (num_col,)))
    model.add(Dense(10,activation= 'relu')) 
    model.add(Dense(10,activation= 'relu')) 
    model.add(Dense(1))
    
    
    #compiling the model
    model.compile(optimizer = 'adam', loss = "mean_squared_error")
    
    #fitting the model
    
    model.fit(x_train,y_train, epochs = 50, verbose =2)
    
    
    #evaluate model
    
    y_pred = model.predict(x_test)
    
    
    mse = mean_squared_error(y_test,y_pred)
    my_mse.append(mse)
    
    
mean_of_mse = np.mean(my_mse)
std_of_mse = np.std(my_mse)

```

```python
#Printing out the 50 mse values added to the list my_mse

print(my_mse)
```

```python


print(f"mean of mean squared error is :  {mean_of_mse}")
print(f"Standard deviation of mean squared error is :  {std_of_mse}")
```
